Bar plot

```python
def grade(score):
  if score >=90:
    return 'A'
  elif score >=70:
    return 'B'
  elif score >=50:
    return 'C'
  else:
    return 'F'

df['grade']=df['exam_score'].apply(grade)
df.head()
```

```python
sns.barplot(data=df,x='grade',y='exam_score',estimator='median')
plt.title("performance of year exam")
plt.tight_layout()
plt.show()
```
