```python
weekly_sales = {}

n = int(input("How many items? "))

for i in range(n):
    item = input("Enter item name: ")
    quantity = int(input("Enter quantity sold: "))

    weekly_sales[item] = quantity

print("\nWeekly Sales Summary")

for item in weekly_sales:
    print(item, ":", weekly_sales[item])
```

#create a 5*5 matrix of random integer (1-90)
#find mean of value row and column wise
#replace all even no with -1
#sort the array row and column wise
#find row and column wise seperetly
#count the no of -ve in each column



```python
import numpy as np
arr=np.random.randint(1,91,(5,5))
print("\n matrix")

print(arr)

row_mean=np.mean(arr,axis=1)
col_mean=np.mean(arr,axis=0)
print("\n row wise mean")

print(row_mean)
print("\n col wise mean")
print(col_mean)

arr[arr%2==0]=-1
print("\n even")
print(arr)

row_sorted=np.sort(arr,axis=1)
print("\n row wise sprted")
print(row_sorted)

col_sorted=np.sort(arr,axis=0)
print("\n col wise sprted")
print(col_sorted)

row_sum=np.sum(arr,axis=1)
print("\n row wise sum")
print(row_sum)

col_sum=np.sum(arr,axis=1)
print("\n col wise sum")
print(col_sum)

negative_count=np.sum(arr<0,axis=0)
print("\n negative")
print(negative_count)


```

```python
import numpy as np
create_stock=np.random.randint(20,81,(5,5))
print("\n create stock")
print(create_stock)

create_stock_demand=np.random.randint(20,81,(5,5))
print("\n create stock_demand")
print(create_stock_demand)


print("\n remaing stock")
print(create_stock - create_stock_demand)

print("\n out of strock")
out_ofstock=np.where(create_stock-create_stock_demand<0,0,create_stock-create_stock_demand)
print(out_ofstock)

print("\n total sortage")
total_sortage=np.sum(np.where(create_stock-create_stock_demand<0,1,0),axis=0)
print(total_sortage)

# products with zero stock
zero_stock=np.sum(np.where(create_stock-create_stock_demand==0,1,0),axis=0)
print("\n products with zero stock")
print(zero_stock)

# sort each row based on remaining stock
print("\n sort each row based on remaining stock")
print(np.sort(out_ofstock,axis=1))





```
