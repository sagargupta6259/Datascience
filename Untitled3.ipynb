```python
z = input()
is_palin = True
i = z[0]
j = z[len(z)-1]
while(i<j):
  if(i!=j):
    is_palin = False
    break
  i+=1
  j-=1


if(i==j):
  print(is_palin)
```

```python
lst=[1,2,3,4,5,"hello",'c',[10,7]]
print(lst)

# append()
lst.append([2,3,4])
print(lst)

# extend()
lst.extend([2,3,4])
print(lst)

# insert()
lst.insert(2,"value")
print(lst)

# pop last value remove
lst.pop()
print(lst)

# remove  fist occurence remove
lst.remove(2)
print(lst)

# indix
print(lst.index(2))

# count
print(lst.count(3))

# sort
lst = [1, 5, 3, 4, 52]
lst.sort()
print(lst)
```
